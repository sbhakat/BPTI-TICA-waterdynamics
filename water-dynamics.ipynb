# BPTI different M states

See Table S2 in the paper entitled 'Transient access to the protein interior: Simulation vs NMR' by Persson and Halle to see the definition of M1 state.

FW: Distance of first closest water molecule to the amide NH.
SW: Distance of second closest water molecule to the amide NH.

In this case we are focusing on Ile18NH.

We will focus on M1 state as it is observable in NMR experiments.

Local fluctuation will allow FW within 0.26 m of amide NH (Ile18NH). See the paper 'How amide hydrogens exchage in native proteins' for more details
